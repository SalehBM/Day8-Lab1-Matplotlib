## Q1: Create two charts using list and dataframe

```python
plt.subplot(121)
plt.plot([1,2,3,4,5])
plt.subplot(122)
plt.bar(pd.Series([3, 4, 8, 10, 6, 5, 4]), 10)
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(np.random.randint(5, 10, size = 5), np.full(5, 0.05));
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
fig, ax = plt.subplots(figsize=(10, 6))
plt.scatter(np.random.randint(1, 25, size = 50), np.random.randint(1, 100, size=50), marker = "o", alpha = 0.4)
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
ax.set_title("Scatter Chart", loc="left", c="r", size = 14);
ax.set_xlabel("x", size = 16)
ax.set_ylabel("y", size = 16)
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.figure(figsize=(10,6))
plt.plot(np.arange(1, 18), c="r", label="Line 1")
plt.plot(np.arange(5, 20), c="b", label="Line 2")
plt.plot(np.arange(1, 35, 3), c="g", label="Line 3")
plt.legend(loc="upper left")
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize= (10, 6))
plt.subplot(131)
plt.bar(np.arange(0, 10), np.arange(0, 10))
plt.subplot(132)
plt.hist(np.arange(10, 15), np.arange(1, 15))
plt.subplot(133)
```
